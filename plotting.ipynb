### Box Plot
Using Pandas and PyPlot for creating Box Plots from CSV Files 
* [PyPlot](https://plot.ly/pandas/box-plots/)

One can store data in data.csv and run the following Scripts!

```python
import string
import pandas as pd
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot


#file to read in
infile = 'data.csv'

#pull data into NumPy dataframe
df = pd.read_csv(infile, sep=';', na_values='.')

data = []

for col in df.columns:
    data.append(go.Box(y=df[col], name=col, showlegend=False))

#print plot
init_notebook_mode(connected=True)

iplot(data)
```

```python
import string
import pandas as pd
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

    
N = 100
y_vals = {}
for letter in list(string.ascii_uppercase):
    y_vals[letter] = np.random.randn(N) + (3 * np.random.randn())

df = pd.DataFrame(y_vals)
df.head()

data = []

for col in df.columns:
    data.append(go.Box(y=df[col], name=col, showlegend=False))

# data.append(
#     go.Scatter(x=df.columns, y=df.mean(), mode='lines', name='mean'))

# IPython notebook
init_notebook_mode(connected=True)
iplot(data, filename='box-plot.html')
```
